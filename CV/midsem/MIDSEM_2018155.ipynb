```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Question 4: LBP Map
def LBP_Map(data):
    data1 = np.full(data.shape,0,dtype = np.uint8) # Zero Padding
    d = {0:(-1,-1),1:(-1,0),2:(-1,1),3:(0,1),4:(1,1),5:(1,0),6:(1,-1),7:(0,-1)}
    eps = 1e-15
    for i in range(1,data.shape[0]-1):
        for j in range(1,data.shape[1]-1):
            for v in range(8):
                ic = i + d[v][0]; jc = j + d[v][1]
                data1[i,j] += round((min(data[ic,jc],data[i,j]) + eps)/(max(data[ic,jc],data[i,j]) + eps))*2**(v)
    return data1 

# question 3: Saliency Map
def Saliency(r):
    sal = np.full(r.shape[:2],0)
    factor = r.shape[0]*r.shape[1]
    for i in range(r.shape[0]):
        for j in range(r.shape[1]):
            sal[i,j] = np.linalg.norm(r - r[i,j],ord = np.inf,axis = 2).sum()
            print(i,j)
    return sal/factor
        

#question 1: 
def q1_1(data):
    hist = np.zeros((256))
    for i in range(256):
        hist[i] += (data == i).sum()
    hist = hist/hist.sum()
    thre = -1 ; val = np.inf
    for i  in range(1,256):
        w1 = hist[:i].sum()
        w2 = hist[i:].sum()
        s1_mean = np.array([ii*hist[ii]/w1 for ii in range(i)]).sum()
        s2_mean = np.array([ii*hist[ii]/w2 for ii in range(i,256)]).sum()
        s1 = np.square([ii - s1_mean for ii in range(i)]).sum()
        s2 = np.square([ii - s2_mean for ii in range(i,256)]).sum()
        if (s1 + s2 < val):
                val = s1+s2
                thre = i
    mask = data >= thre
    print(thre)
    return mask
        
```

```python
rgb_img = Image.open('iiitd2.png')
rgb_data = np.asarray(rgb_img)
grey_img = Image.open('iiitd1.png')
grey_data = np.asarray(grey_img)
```
